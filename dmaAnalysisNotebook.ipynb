# Automated DMA Analysis
This notebook demonstrates how to use analysisDMA to analyze several common types of DMA experiments: strain sweep, temperature sweep (multiple frequencies), creep, and stress relaxation. It accepts raw data files from a TA Q800 DMA and will output relevant values to each experiment in a dataframe (e.g. glass transition temperature) as well as plot the relevant signals.

```python
import analysisDMA
from analysisDMA import *
```

# Strain sweep
Returns the strain at the end of the linear viscoelastic regime (LVE).

```python
strain1 = strain('Strain_1.txt')
strain1.strain_plot()
display(strain1.df_analysis)
```

# Creep
Returns the elastic strain, creep strain, elastic recovery, creep recovery and permenant deformation. Will also work for creep experiments with multiple cycles.

```python
creep1 = creep('Creep_3.txt')
creep1.creep_plot()
display(creep1.df_analysis)
```

# Temperature sweep (one or multifrequency)
Returns the glass transition for each frequency measured determined by both Tan Delta and Loss Modulus

```python
temp1 = tempSweep('Temp_Sweep_1.txt')
temp1.temp_sweep_plot()
display(temp1.df_analysis)
```

# Stress relaxation
Returns the total stress relaxed over the course of the measurement (total time). Will also return the Charateristic Relaxation Time (tau) which is initial stress/e, if the experiment achieved this level of stress relaxation, otherwise its NaN

```python
sr1 = stressRelax('Stress_relax_2.txt')
sr1.stress_relax_plot()
display(sr1.df_analysis)
```
